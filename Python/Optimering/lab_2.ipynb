```python
import scipy.optimize as lp
import numpy as np


A_eq = np.array([
    [1, 1, 1, 0, 0, 0, 0, 0, 0],  # Supply: Älmhult
    [0, 0, 0, 0, 0, 0, 1, 0, 0],  # Supply: Markaryd
    [0, 0, 0, 0, 1, 0, 1, 0, 0],  # Demand: Ljungby
    [0, 0, 0, 0, 0, 0, 0, 1, 0],  # Demand: Alvesta
    [0, 0, 1, 0, 0, 1, 0, 0, 1],  # Demand: Växjö
])

A_ub = np.eye(9)

b_eq = np.array([2500, 1000, 1000, 500, 2000])

b_ub = np.array([1000, 1000, 1000, 1000, 500, 1000, 2000, 500, 2000])

c = np.array([18, 24, 62, 29, 31, 46, 51, 42, 20])

result = lp.linprog(c, A_eq=A_eq, b_eq=b_eq, A_ub=A_ub, b_ub=b_ub)

result_list = ["Älmhult to Liatorp", "Älmhult to Osby", "Älmhult to Växjö", "Osby to Markaryd", "Liatorp to Ljungby", "Liatorp to Växjö", "Markaryd to Ljungby", "Ljungby to Alvesta", "Alvesta to Vaxjö"]

print(f"The optimal solution gives a cost of {result.fun} and:\n")

for i in range(len(result_list)):
    print(f"\tFlow of {result.x[i]} from {result_list[i]}")
```

## Steepest Descent

```python

def func_f(x):
    return (x[0] + 1)**2 - x[0]*x[1] + 3*(x[1] - 5)**2


def func_df(x):
    return np.array([2*x[0] + 2 - x[1], -x[0] + 6*x[1] - 30])


def func_w(t, x):
    return x - t*func_df(x)


def steepest_descent(x):
    t, counter = 1, 0

    history = [x]

    while not np.sqrt(func_df(x).T @ func_df(x)) < 1e-3:      # Check if gradient is within tolerance

        if func_f(func_w(t, x)) <= func_f(x) - 0.2*t*func_df(x).T @ func_df(x):
            while func_f(func_w(t, x)) <= func_f(x) - 0.2*t*func_df(x).T @ func_df(x):
                t = 2*t
            t = t/2
            counter += 1
        else:
            while func_f(func_w(t, x)) > func_f(x) - 0.2*t*func_df(x).T @ func_df(x):
                t = t/2
            counter += 1

        x = func_w(t, x)
        history += [x]
    return x, counter, history


def abs_error(xi, xj):
    x = xi - xj
    return np.linalg.norm(x)


min = np.array([28-5*(58/11), 58/11])   # Solved manually

point = np.array([1, 1])

print(f"Starting point (x, y): {(point[0], point[1])}")

point, counter, history = steepest_descent(point)

print(f"Number of iterations until ||df(x_k)|| < 10^-3: {counter}")
print(f"Absolute error: {abs_error(point, min):.3}")
```

## Newton Method

```python
def newton_method(i_h, x):
    counter = 0

    history =[x]
    while not np.sqrt(func_df(x).T @ func_df(x)) < 1e-3:
        x = x - i_h @ func_df(x)
        counter += 1
        history += [x]
    return x, counter, history


point = np.array([1, 1])

print(f"Starting point (x, y): {(point[0], point[1])}")

hessian = np.array([
    [2, -1],
    [-1, 6]
])

if np.linalg.det(hessian) == 0:
    print("Non-invertible hessian matrix!")
    exit()
else:
    inv_hessian = np.linalg.inv(hessian)

point, counter, history = newton_method(inv_hessian, point)

print(f"Number of iterations until ||df(x_k)|| < 10^-3: {counter}")
print(f"Absolute error: {abs_error(point, min):.3}")
```

### Task 1

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm

def rosenbrock(x):
    return (1-x[0])**2 + 100*(x[1]-x[0]**2)**2


def rosenbrock_df(x):
    return np.array([-2*(1-x[0]) - 400*x[0]*(x[1]-x[0]**2), 200*(x[1]-x[0]**2)])


def func_w_rosenbrock(t, x):
    return x - t*rosenbrock_df(x)


def steepest_descent_rosenbrock(x):
    t = 1

    history = [x]

    while not np.linalg.norm(rosenbrock_df(x)) < 1e-4:      # Check if gradient is within tolerance

        if rosenbrock(func_w_rosenbrock(t, x)) <= rosenbrock(x) - 0.2*t*rosenbrock_df(x).T @ rosenbrock_df(x):
            while rosenbrock(func_w_rosenbrock(t, x)) <= rosenbrock(x) - 0.2*t*rosenbrock_df(x).T @ rosenbrock_df(x):

                t = 2*t

            t = t/2

        else:
            while rosenbrock(func_w_rosenbrock(t, x)) > rosenbrock(x) - 0.2*t*rosenbrock_df(x).T @ rosenbrock_df(x):

                t = t/2

        x = func_w_rosenbrock(t, x)
        history += [x]
        
    return x, history


def newton_method_rosenbrock(x):
    history = [x]
    
    while np.linalg.norm(rosenbrock_df(x)) >= 1e-4:
        
        hessian = np.array([
            [2 - 400 * x[1] + 1200 * x[0]**2, -400 * x[0]],
            [-400 * x[0], 200]
        ])
        
        det_h = hessian[0, 0] * hessian[1, 1] - hessian[0, 1] * hessian[1, 0]
        
        inv_h = (1 / det_h) * np.array([
            [hessian[1, 1], -hessian[0, 1]],
            [-hessian[1, 0], hessian[0, 0]]
        ])
        
        x = x - inv_h @ rosenbrock_df(x)
        
        history.append(x)
    
    return x, history


def calculate_trajectory_length(history):
    trajectory_length = 0
    for i in range(len(history) - 1):
        dist = np.linalg.norm(history[i+1] - history[i])
        trajectory_length += dist
    return trajectory_length


x = np.linspace(-2, 2, 1000)
y = np.linspace(-1, 3, 1000)

[X, Y] = np.meshgrid(x, y)

Z = rosenbrock([X, Y])

plt.figure(1)
plt.contourf(X, Y, Z, 50, cmap=cm.jet)
plt.title('Rosenbrock')
plt.colorbar()

plt.figure(2)
plt.contourf(X, Y, np.log(Z), 50, cmap=cm.jet)
plt.title('Rosenbrock (log)')
plt.colorbar()
```

### Task 2

```python
x0 = np.array([-1, 1.5])

p, history = steepest_descent_rosenbrock(x0)

plt.figure(3)
plt.contourf(X, Y, np.log(Z), 50, cmap=cm.jet)
plt.title("Steepest Descent")
plt.colorbar()
px = [c[0] for c in history]
py = [c[1] for c in history]
plt.plot(px, py, '-', label='Steepest Descent Path')
plt.plot(px[0], py[0], 'o', color='black', label='Start Point')
plt.plot(px[-1], py[-1], 'o', color='white', label='End Point')
plt.xlim(-2, 2)
plt.ylim(-1, 3)
plt.legend()
plt.show()

print(f"The minimum of the rosenbrock function lies at (x, y): {(round(px[-1], 5), round(py[-1], 5))}")
print(f"The trajectory of the steepest descent is {calculate_trajectory_length(history)} long")
print(f"with {len(history)} iterations")
```

### Task 3

```python
x0 = np.array([-1, 1.5])

p, history = newton_method_rosenbrock(x0)

plt.figure(4)
plt.contourf(X, Y, np.log(Z), 50, cmap=cm.jet)
plt.title("Newton Method")
plt.colorbar()
px = [c[0] for c in history]
py = [c[1] for c in history]
plt.plot(px, py, '-', label='Newton Method Path')
plt.plot(px[0], py[0], 'o', color='black', label='Start Point')
plt.plot(px[-1], py[-1], 'o', color='white', label='End Point')
plt.xlim(-2, 2)
plt.ylim(-1, 3)
plt.legend()
plt.show()

print(f"The minimum of the rosenbrock function lies at (x, y): {(round(px[-1], 5), round(py[-1], 5))}")
print(f"The trajectory of the newton method is {calculate_trajectory_length(history)} long")
print(f"with {len(history)} iterations")
```
