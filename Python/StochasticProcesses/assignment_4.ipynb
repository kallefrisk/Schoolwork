## Problem 1.

### One simulation of the system

```python
import random

lamd1 = 1
lamd2 = 2

time = 0
simulations = 100

# Does the system function?
functioning = True
# Do the components function?
comp1, comp2 = True, True
# To know if any component has broken
first = True

while functioning:
    comp_1_time = random.expovariate(lamd1)
    comp_2_time = random.expovariate(lamd2)

    if comp_1_time <= comp_2_time:
        time += comp_1_time
        comp1 = False
        if first:
            first_broken = 1
    else:
        time += comp_2_time
        comp2 = False
        if first:
            first_broken = 2
    
    functioning = False if not comp1 and not comp2 else None


print('Time for system to fail: ', time, '\nFirst component to break: ', first_broken)
```

### Multiple simulations of the system

```python
import random

lamd1 = 1
lamd2 = 2

total_time = 0
simulations = 10000


comp_1_first = 0
comp_2_first = 0


for _ in range(simulations):
    time = 0
    # Does the system function?
    functioning = True
    # Do the components function?
    comp1, comp2 = True, True
    # To know if any component has broken
    first = True
    while functioning:
        comp_1_time = random.expovariate(lamd1)
        comp_2_time = random.expovariate(lamd2)

        if comp_1_time <= comp_2_time:
            time += comp_1_time
            comp1 = False
            if first:
                comp_1_first += 1
        else:
            time += comp_2_time
            comp2 = False
            if first:
                comp_2_first += 1
        
        functioning = False if not comp1 and not comp2 else None
    
    total_time += time
    

print('Average time to failure: ', total_time/simulations, '\nFraction of time component 1 broke first: ', comp_1_first/simulations)
```

## Problem 3 b)

### Multiple simulations of the system

```python
import random

lamd1 = 1
lamd2 = 2

# ----------------------------------------------------------------------------------------------------------------------------------------------------
# System a
# ----------------------------------------------------------------------------------------------------------------------------------------------------

total_time = 0
simulations = 10000

for _ in range(simulations):
    
    time = 0
    # Does the system function?
    functioning = True
    # Do the components function?
    comp1a, comp1b, comp2a, comp2b = True, True, True, True
    # To know if any component has broken
    first = True
    while functioning:
        comp_1a_time = random.expovariate(lamd1)
        comp_1b_time = random.expovariate(lamd1)
        comp_2a_time = random.expovariate(lamd2)
        comp_2b_time = random.expovariate(lamd2)

        shortest_time = min(comp_1a_time, comp_1b_time, comp_2a_time, comp_2b_time)

        if comp_1a_time == shortest_time:
            time += comp_1a_time
            comp1a = False
        elif comp_1b_time == shortest_time:
            time += comp_1b_time
            comp1b = False
        elif comp_2a_time == shortest_time:
            time += comp_1a_time
            comp2a = False
        else:
            time += comp_2b_time
            comp2b = False
        
        functioning = False if (not comp1a and not comp2a) or (not comp1b and not comp2b) else None
    total_time += time


print('Average time for system a to fail: ', total_time/simulations)


# ----------------------------------------------------------------------------------------------------------------------------------------------------
# System b
# ----------------------------------------------------------------------------------------------------------------------------------------------------

total_time = 0

for _ in range(simulations):
    
    time = 0
    # Does the system function?
    functioning = True
    # Do the components function?
    comp1a, comp1b, comp2a, comp2b = True, True, True, True
    # To know if any component has broken
    first = True
    while functioning:
        comp_1a_time = random.expovariate(lamd1)
        comp_1b_time = random.expovariate(lamd1)
        comp_2a_time = random.expovariate(lamd2)
        comp_2b_time = random.expovariate(lamd2)

        shortest_time = min(comp_1a_time, comp_1b_time, comp_2a_time, comp_2b_time)

        if comp_1a_time == shortest_time:
            time += comp_1a_time
            comp1a = False
        elif comp_1b_time == shortest_time:
            time += comp_1b_time
            comp1b = False
        elif comp_2a_time == shortest_time:
            time += comp_1a_time
            comp2a = False
        else:
            time += comp_2b_time
            comp2b = False
        
        functioning = False if (not comp1a and not comp1b) or (not comp1a and not comp2b) else None
    total_time += time


print('Average time for system b to fail: ', total_time/simulations)
```
