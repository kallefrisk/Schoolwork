## Computer lab 6: Finite Element Method for Helmholtz problems

In this computer lab, we continue revising wave problems posed in frequency domain. For convenience, we work with solutions using complex numbers, however we ultimately recover solutions that are real by taking the real part of the resulting complex function.

Let us start by considering the following wave equation with a harmonic source term:

Find $w$ such that 
$$
\begin{array}{rl}

\end{array}
$$
subject to given boundary conditions for $w(x,t)$, and

In the case where sources are harmonic with frequency $\omega$, it makes sense to assume a harmonic solution: we propose to use $w(x,t)=\text{Re} \left(u(x)e^{-i\omega t}\right)$.

## Helmholtz source problem in 1D
For a given frequency $\omega$ and source $f=1$ we are interested in finding the unique wave $u$ satisfying a Helmholtz equation subject to given boundary conditions.

Consider the following source problems:

### 1) Case with fixed ends (Dirichlet)
Find $u$ such that 
$$
\begin{array}{rl}
-u''-\omega^2 u =&\!\!\!\! 1,\,\,x\in(0,1) \\
u(0)=&\!\!\!\!0,  \\
u(1)=&\!\!\!\!0.
\end{array}
$$

### 2) Case with outgoing condition (Robin)
Find $u$ such that 
$$
\begin{array}{rl}
-u''-\omega^2 u =&\!\!\!\! 1,\,\,x\in(0,1) \\
u(0)=&\!\!\!\!0,  \\
u'(1)=&\!\!\!\!i\omega \,u(1),
\end{array}
$$

## $L_2$-errors in FE approximations

Let $u$ be an exact solution to a particular Helmholtz problem, $u_h$ be a FE approximation and $P_h u$ be the L2-projection to the exact solution in the basis $\{\varphi_i\}_{i=1}^N$. 

For the FE representation written as $$u_h=\sum_j\xi_j\varphi_j,$$
and the L2-projection given as $$P_h u=\sum_j\eta_j\varphi_j,$$
we compute the $L_2$-error $E_h$ in FE approximations as
$$E_h:=\|P_hu-u_h\|=\left(\int_\Omega |P_hu-u_h|^2 dx\right)^{1/2}.$$

**Remark:** for a complex number $z$, the value $\bar z$ denotes its conjugated. Squared magnitudes are computed as $|z|^2=\bar z z$.

## Implementation for source problems in 1D:

Below we provide an implementation for the FE approximation of the Helmholtz case (2) where we reuse the scripts introduced in earlier computer labs. 
Study the code and make sure that you understand what has been done. You should be able to understand how boundary conditions are implemented in the code. Particularly, modify the number of nodes `N` and study the resulting FE approximations $u_h$ and error $E_h$.

```python
# Example FEM Helmholtz 1D

import numpy as np
import matplotlib.pyplot as plt
import mpmath
import scipy.integrate

print('Helmholtz with outgoing condition in 1D')

# Source
def f(x):
    return 1.0

# Exact solution
def ue(x,w):
    Bc=np.exp(1j*w)/(2.0*w*w)
    Cc=-1.0/(w*w)
    Ac=-Bc-Cc
    return (Ac*np.exp(1j*w*x)+Bc*np.exp(-1j*w*x)+Cc)

# Number of nodes
N=10

nodes    = np.linspace(0,1,N)
elements = np.zeros((N-1,2)).astype('int')

for i in range(N-1):
    elements[i][0] = i
    elements[i][1] = i+1

# X-values for plotting purposes
x = np.linspace(nodes[0], nodes[-1], 201)

N = len(nodes)

w = 10.0
w2 = w*w

# Allocation of Global matrix and vectors
A = np.zeros((N,N), dtype=complex)
M = np.zeros((N,N), dtype=complex)
S = np.zeros((N,N), dtype=complex)
b = np.zeros(N, dtype=complex)
z = np.zeros(N, dtype=complex)

# Local 1D matrices
Al = np.array([[1,-1], [-1,1]])
Ml = np.array([[2, 1], [1, 2]])

# FE assembly
for k in range(len(elements)):
    i = elements[k][0]

    x0 = nodes[elements[k][0]]
    x1 = nodes[elements[k][1]]
    xm = (x0+x1)/2

    h = x1-x0

    A[i][i]     += Al[0][0]/h
    A[i][i+1]   += Al[0][1]/h
    A[i+1][i]   += Al[1][0]/h
    A[i+1][i+1] += Al[1][1]/h

    M[i][i]     += (h/6)*Ml[0][0]
    M[i][i+1]   += (h/6)*Ml[0][1]
    M[i+1][i]   += (h/6)*Ml[1][0]
    M[i+1][i+1] += (h/6)*Ml[1][1]

    # Simpson rule, FEM load vector
    b[i]        += (h/6)*(1.0*f(x0)+4*0.5*f(xm)+0.0*f(x1))
    b[i+1]      += (h/6)*(0.0*f(x0)+4*0.5*f(xm)+1.0*f(x1))
    
    # Simpson rule, Best approximation
    z[i]        += (h/6)*(1.0*ue(x0,w)+4*0.5*ue(xm,w)+0.0*ue(x1,w))
    z[i+1]      += (h/6)*(0.0*ue(x0,w)+4*0.5*ue(xm,w)+1.0*ue(x1,w))

# System matrix for FEM
S = A-w2*M

# Dirichlet boundary conditions on x=0.0
for j in range(N):
    S[0][j] = 0.
    S[j][0] = 0.
S[0][0] = 1.
b[0]    = 0

# Outgoing boundary condition on x=1.0
S[-1][-1] += -1j*w

# Solving the linear system: FEM
U = np.linalg.solve(S, b)

# Solving the linear system: Best approximation
p = np.linalg.solve(M, z)

# Computing interpolation at nodal values
u = ue(x,w)

# residual
y   = p-U
yc  = np.conjugate(y)

# Computing and output L2-error
Eh = np.sqrt(np.dot(yc.T ,np.dot(M, y)))
print ('FE L2 error (N,err):')
print ('',N,',', Eh.real)

# Plotting routine
plt.plot(nodes,np.zeros(len(nodes)),'b.')
plt.plot(x,u.real,'k-')
plt.plot(nodes,U.real,'r-', label='FEM')
plt.plot(nodes,p.real,'b-', label='best approx')
plt.legend()

plt.show()
```

## Implementation for eigenvalue problems in 1D:

Below we provide an implementation for the FE approximation of the Helmholtz case (3).

```python
# Example FEM 1D

import numpy as np
import matplotlib.pyplot as plt
import mpmath
import scipy.integrate

print('Helmholtz eigenvalue problem in 1D')

# Exact eigenfunctions
def ue(x,m):
    return np.sin(m*np.pi*x)

# Number of nodes
N=10

nodes    = np.linspace(0.,1.,N)
elements = np.zeros((N-1,2)).astype('int')

for i in range(N-1):
    elements[i][0] = i
    elements[i][1] = i+1

x = np.linspace(nodes[0], nodes[-1], 201)

N = len(nodes)

# Global matrix and vectors
A = np.zeros((N,N))#, dtype=np.complex)
M = np.zeros((N,N))#, dtype=np.complex)
y = np.zeros(N)

# Local matrix and vector
Al = np.array([[1.,-1.], [-1.,1.]])
Ml = np.array([[2., 1.], [1., 2.]])

for k in range(len(elements)):
    i  = elements[k][0]
    
    x0 = nodes[elements[k][0]]
    x1 = nodes[elements[k][1]]
    xm = (x0+x1)/2.

    h = x1-x0

    A[i][i]     += Al[0][0]/h
    A[i][i+1]   += Al[0][1]/h
    A[i+1][i]   += Al[1][0]/h
    A[i+1][i+1] += Al[1][1]/h

    M[i][i]     += (h/6.)*Ml[0][0]
    M[i][i+1]   += (h/6.)*Ml[0][1]
    M[i+1][i]   += (h/6.)*Ml[1][0]
    M[i+1][i+1] += (h/6.)*Ml[1][1]

# Dirichlet boundary conditions applied on both A and M
for j in range(N):
    A[0][j] = 0.
    A[j][0] = 0.
    A[N-1][j] = 0.
    A[j][N-1] = 0.
    M[0][j] = 0.
    M[j][0] = 0.
    M[N-1][j] = 0.
    M[j][N-1] = 0.
    
M[0][0] = 1.
M[-1][-1] = 1.

w2, U = scipy.linalg.eigh(A, M,  eigvals_only=False) #, eigvals_only=False, subset_by_index=[0, 1, 2])
print('Frequencies: ',np.sqrt(w2))

for m in range(len(w2)):
    y = A.dot(U[:,m])-w2[m]*M.dot(U[:,m])
    # print(scipy.linalg.norm(y))

# index m=1,2,... denotes the ordering of eigenpairs
m=1 # First two eigenvalues are spurious due to reduction of linear system

# normalization factor
Umax = np.max(np.abs(U[:,m+1].real))
if Umax > 1.0e-12:
    # Fixing the sign and normalization
    Um = np.sign(U[2,m+1])*U[:,m+1].real/Umax 
u = ue(x,m)
plt.plot(nodes,np.zeros(len(nodes)),'b.')
plt.plot(x,u.real,'k-')
plt.plot(nodes,Um,'r-')

plt.show()
```

### Remark on spurious eigenpairs:
Notice that the first two eigenvalues of the generalized eigenvalue problem are zero.
These eigenvalues are spurious: the reason these spurious eigenpairs appear is because of the way we choose to impose Dirichlet boundary conditions. One effective way to avoid spurious eigeinpairs in these sort of eigenproblems is to reduce the dimension of the resulting matrix system.

## Helmholtz source problems in 2D

In this section we deal with a Helmholtz equation of the form:
$$
-\Delta u-\omega^2 m(x) u =f(x),
$$
subject to given boundary conditions.

First, we use the FE machinery developed in the lectures and in Computer Lab 5. Particularly, we use the definition of the matrix $R$ with contribution from elements having and edge on the boundary of $\Omega$.
Their elements are computed as
$$
R_{ij}=\int_{\Gamma}\varphi_j\varphi_i\,dx.
$$
This matrix is used for enforcing outgoing conditions.

### Dirichlet, Neumann and outgoing boundary conditions

We move on by illustrating the application of different boundary conditions in Helmholtz problems. We choose **Dirichlet**, **Newmann** and **outgoing** boundary conditions. 

We continue to perform computations in the squared mesh $\Omega:=[0,L]^2$.
The boundary is split in the following way:
Let $\Gamma_D:=\{(x,y):x=0\,\,\,\text{and}\,\,\,0\leq y \leq L\}$ 
be the left edge of $\Omega$.
Additionally, let $\Gamma_N$ be the union of the top, bottom and right faces of $\Omega$. In this way $\partial\Omega=\Gamma_D\cup\Gamma_N$.

In the following task, we consider the following Helmholtz problems:

Find $u$ such that:
$$
\begin{array}{rll}
-\Delta u -\omega^2 u=& \!\!\! 0, & x\in\Omega \\
u=&  \!\!\! 1, &x\in \Gamma_D \,\,\text{(Dirichlet)}.
\end{array}
$$

**Remark:** Notice that with this choice the source of waves is generated in the boundary $\Gamma_D$. Think of the wave $w(x,t)$ being generated in the edge $\Gamma_D$. The forcing function in time is $g(x,t)=1\cdot e^{-i\omega t}$ for $x\in\Gamma_D$.

Assuming that the boundary of the domain $\Gamma$ is split in two sets $\Gamma:=\Gamma_D\cup\Gamma_N$. We may consider an additional boundary condition imposed on $x\in\Gamma_N$:

**a)** $n\cdot\nabla u=0,\,\,\,x\in \Gamma_N$ (Neumann)

**b)** $n\cdot\nabla u=i\omega \,u,\,\,\,x\in \Gamma_N$ (Outgoing).

### Implementation for Helmholtz problems in 2D

In order to perform FE computations for Helmholtz problems in 2D, we can reuse the machinery developed in Computer Lab 5.
In the script below we gather the ideas developed in earlier computer labs and present definitions for the assembly of FE in 2D.

```python
import numpy as np
from math import ceil

import matplotlib.pyplot as plt
from matplotlib import cm
from mpl_toolkits.mplot3d import Axes3D
# %matplotlib widget
# %matplotlib notebook

import scipy.sparse as sp
from scipy.sparse import identity, dok_matrix, spdiags
from scipy.sparse.linalg import spsolve
from scipy.sparse.linalg import factorized

import meshtools as mt
import plottools as pt

def extract_nodes(P, inside_domain):
    """Short function which extract all node indices i
    for which the inside_domain(P[i]) evaluates to True.
    """
    return [ i for i in range(P.shape[0]) if inside_domain(P[i]) ]

def compute_hat_gradients(tri):
    # Compute area
    N0, N1, N2 = tri
    area=abs(0.5*np.cross(N1-N0, N2-N0))

    # Compute b = (1,1,1) x (x_2^1,x_2^2,x_2^3). c is similar
    ones = np.ones(3)
    b = np.cross(tri[:,1], ones)/(2.*area)
    c = np.cross(ones, tri[:,0])/(2.*area)

    return (area, b, c)

def assemble_FEM(P, T, E, n2, f):
    
    # 2 times the mass matrix on a unit triangle
    M_ref = 1.0/12.0*np.array([[2, 1, 1],
                               [1, 2, 1],
                               [1, 1, 2]])
    
    # Create matrix
    n_p = P.shape[0]

    print("Create stiffness matrix of size %d x %d" % (n_p, n_p))
    A   = dok_matrix((n_p, n_p))
    M   = dok_matrix((n_p, n_p))
    R   = dok_matrix((n_p, n_p))
    rhs = np.zeros(n_p)
    
    # Iterate over all triangles
    n_t = T.shape[0]

    print("Assemble matrices over %d elements ..." % n_t)
    for K in range(n_t):
        l2g = T[K]   # Get local to global map
        tri = P[l2g] # Get triangle coordinates
        
        # TODO: Add computation of abar!
        area, b, c = compute_hat_gradients(tri)
        
        # Local element matrix
        N0,N1,N2 = tri 
        area=abs(0.5*np.cross(N1-N0,N2-N0))
        
        cp = (N0+N1+N2)/3.0
        mp =  0.5*np.array([N0+N1, N1+N2, N2+N0])
        b_K = f(tri[:,0], tri[:,1])*area/3
        
        rhs[np.ix_(l2g)] += b_K
        
        M_K = n2(cp[0],cp[1])*M_ref*area
        M[np.ix_(l2g, l2g)] += M_K
        
        A_K = (np.outer(b, b) + np.outer(c, c))*area
        A[np.ix_(l2g, l2g)] += A_K
    
    # Assemble of boundary Mass matrix
    Re  = np.array([[2.,1.], [1.,2.]])
    for e in range(len(E)):
        l2g = E[e]
        x=np.array(P[l2g,0])
        y=np.array(P[l2g,1])

        h = np.sqrt( (x[1]-x[0])**2 + (y[1]-y[0])**2)

        R[np.ix_(l2g, l2g)] += (h/6.)*Re
    
    print("Done.")   
    return A, M, R, rhs
```

### Scattering problem

By demanding that $u$ be an outgoing wave travelling away from the scatterer $\Omega_s$, we arrive at the following scattering problem:

Find $u$ such that
$$
\begin{array}{rl}
-\Delta u-\omega^2 m(x)u= & \!\!\! f(x),\,\,\,x\in \Omega, \\
n\cdot \nabla u= & \!\!\! i\omega\, u,\,\,\,x\in \Gamma,
\end{array}
$$
for $\omega$ and $f$ known.
Additionally, we have assumed that $\Omega$ is a compact domain containing the scatterer $\Omega_s$.

In the following script we present a FE implementation of the scattering problem. The example employs $\Omega=[0,1]^2$ and $\Omega_s:=\{(x,y):|x-0.5| < 0.1\,\,\,\text{and}\,\,\,|y-0.5| < 0.1\}$.

```python
# Number of subdivision in each spatial direction
N = 51

# Mesh size
L = 1.
h = L/N

# Mesh and coordinates
points, triangles, edges = mt.rectangle_tria((0,0), (L,L), N, N)
X = points[:, 0]
Y = points[:, 1]

Np = len(X)

w=30.0
w2=w*w

# Refractive index function
def m(x, y):
    z=1.0
    if np.abs(x-0.5) < 0.1 and np.abs(y-0.5) < 0.1:
        z = 4.0
    return z

# Source
def f(x, y):
    z=np.zeros(3)
    for i in range(3):
        if np.abs(x[i]-0.5) < 0.1 and np.abs(y[i]-0.5) < 0.1:
            z[i] = w*w*(m(x[i],y[i])-1.0)*np.exp(1j*w*x[i])
    return z


# Compute FE matrices and load vector
A,M,R,b = assemble_FEM (points, triangles, edges, m,f)

S = A-w2*M-1j*w*R
    
S = S.tocsr() # Converting in CSR sparse format

U=spsolve(S, b)

pt.plot2D(X, Y, U.real, triangles=triangles, title="$u(x,y)$")
```
