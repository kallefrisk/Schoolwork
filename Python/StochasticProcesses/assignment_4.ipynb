## Problem 1.

### One simulation of the l system

```python
import random

lamd1 = 1
lamd2 = 2

time = 0
simulations = 100

# Does the system function?
functioning = True
# Do the components function?
comp1, comp2 = True, True
# To know if any component has broken
first = True

while functioning:
    comp_1_time = random.expovariate(lamd1)
    comp_2_time = random.expovariate(lamd2)

    if comp_1_time <= comp_2_time:
        time += comp_1_time
        comp1 = False
        if first:
            first_broken = 1
    else:
        time += comp_2_time
        comp2 = False
        if first:
            first_broken = 2
    
    functioning = False if not comp1 and not comp2 else True


print('Time for system to fail: ', time, '\nFirst component to break: ', first_broken)
```

### Multiple simulations of the parallel system

```python
import random

lamd1 = 1
lamd2 = 2

total_time = 0
simulations = 10000


comp_1_first = 0
comp_2_first = 0


for _ in range(simulations):
    time = 0
    # Does the system function?
    functioning = True
    # Do the components function?
    comp1, comp2 = True, True
    # To know if any component has broken
    first = True
    while functioning:
        comp_1_time = random.expovariate(lamd1)
        comp_2_time = random.expovariate(lamd2)

        if comp_1_time <= comp_2_time:
            time += comp_1_time
            comp1 = False
            if first:
                comp_1_first += 1
                first = False
        else:
            time += comp_2_time
            comp2 = False
            if first:
                comp_2_first += 1
                first = False
        
        functioning = False if not comp1 and not comp2 else True
    
    total_time += time
    

print(f'Average time to failure: \t{total_time/simulations}')
print(f'Theoretical lifetime: \t\t{1/lamd1 + 1/lamd2 - 1/(lamd1 + lamd2)}')
print(f'Fraction of time component 1 broke first: \t{comp_1_first/simulations}')
print(f'Theoretical fraction of time 1 broke first: \t{lamd1/(lamd1 + lamd2)}')
```

## Problem 3 a)

### One simulation of both the systems

```python
import random

lamd1 = 1
lamd2 = 2

# ----------------------------------------------------------------------------------------------------------------------------------------------------
# System a
# ----------------------------------------------------------------------------------------------------------------------------------------------------

time = 0

# Does the system function?
functioning = True
# Do the components function?
comp1a, comp1b, comp2a, comp2b = True, True, True, True

while functioning:
    comp_1a_time = random.expovariate(lamd1)
    comp_1b_time = random.expovariate(lamd1)
    comp_2a_time = random.expovariate(lamd2)
    comp_2b_time = random.expovariate(lamd2)

    shortest_time = min(comp_1a_time, comp_1b_time, comp_2a_time, comp_2b_time)

    if comp_1a_time == shortest_time:
        time += comp_1a_time
        comp1a = False
    elif comp_1b_time == shortest_time:
        time += comp_1b_time
        comp1b = False
    elif comp_2a_time == shortest_time:
        time += comp_2a_time
        comp2a = False
    else:
        time += comp_2b_time
        comp2b = False
    
    functioning = False if (not comp1a and not comp2a) or (not comp1b and not comp2b) else True


print('Time for system to fail: ', time)

# ----------------------------------------------------------------------------------------------------------------------------------------------------
# System b
# ----------------------------------------------------------------------------------------------------------------------------------------------------

time = 0

# Does the system function?
functioning = True
# Do the components function?
comp1a, comp1b, comp2a, comp2b = True, True, True, True

while functioning:
    comp_1a_time = random.expovariate(lamd1)
    comp_1b_time = random.expovariate(lamd1)
    comp_2a_time = random.expovariate(lamd2)
    comp_2b_time = random.expovariate(lamd2)

    shortest_time = min(comp_1a_time, comp_1b_time, comp_2a_time, comp_2b_time)

    if comp_1a_time == shortest_time:
        time += comp_1a_time
        comp1a = False
    elif comp_1b_time == shortest_time:
        time += comp_1b_time
        comp1b = False
    elif comp_2a_time == shortest_time:
        time += comp_2a_time
        comp2a = False
    else:
        time += comp_2b_time
        comp2b = False
    
    functioning = False if (not comp1a and not comp1b) or (not comp2a and not comp2b) else True


print('Time for system to fail: ', time)
```

## Problem 3 b)

### Multiple simulations of both the systems

```python
import random

lamd1 = 1
lamd2 = 2

# ----------------------------------------------------------------------------------------------------------------------------------------------------
# System a (Two series in parallel)
# ----------------------------------------------------------------------------------------------------------------------------------------------------

total_time = 0
simulations = 10000

for _ in range(simulations):
    
    time = 0
    # Does the system function?
    functioning = True
    # Do the components function?
    comp1a, comp1b, comp2a, comp2b = True, True, True, True

    while functioning:
        comp_1a_time = random.expovariate(lamd1)
        comp_1b_time = random.expovariate(lamd1)
        comp_2a_time = random.expovariate(lamd2)
        comp_2b_time = random.expovariate(lamd2)

        shortest_time = min(comp_1a_time, comp_1b_time, comp_2a_time, comp_2b_time)

        if comp_1a_time == shortest_time:
            time += comp_1a_time
            comp1a = False
        elif comp_1b_time == shortest_time:
            time += comp_1b_time
            comp1b = False
        elif comp_2a_time == shortest_time:
            time += comp_2a_time
            comp2a = False
        elif comp_2b_time == shortest_time:
            time += comp_2b_time
            comp2b = False

        functioning = False if (not comp1a or not comp2a) and (not comp1b or not comp2b) else True
    total_time += time


print(f'Average time for system a to fail: {total_time/simulations}')
print(f'Theoretical time to failure: \t{3/2/(lamd1 + lamd2)}')


# ----------------------------------------------------------------------------------------------------------------------------------------------------
# System b (Two parallel in series)
# ----------------------------------------------------------------------------------------------------------------------------------------------------

total_time = 0

for _ in range(simulations):
    
    time = 0
    # Does the system function?
    functioning = True
    # Do the components function?
    comp1a, comp1b, comp2a, comp2b = True, True, True, True

    while functioning:
        comp_1a_time = random.expovariate(lamd1)
        comp_1b_time = random.expovariate(lamd1)
        comp_2a_time = random.expovariate(lamd2)
        comp_2b_time = random.expovariate(lamd2)

        shortest_time = min(comp_1a_time, comp_1b_time, comp_2a_time, comp_2b_time)

        if comp_1a_time == shortest_time:
            time += comp_1a_time
            comp1a = False
        elif comp_1b_time == shortest_time:
            time += comp_1b_time
            comp1b = False
        elif comp_2a_time == shortest_time:
            time += comp_2a_time
            comp2a = False
        else:
            time += comp_2b_time
            comp2b = False
        
        functioning = False if (not comp1a and not comp1b) or (not comp2a and not comp2b) else True
    total_time += time


print(f'Average time for system b to fail: {total_time/simulations}')
print(f'Theoretical time to failure: \t{4/(lamd1 + lamd2) - 2/(2*lamd1 + lamd2) - 2/(lamd1 + 2*lamd2) + 1/2/(lamd1 + lamd2)}')
```
